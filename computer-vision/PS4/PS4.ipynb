```python
import numpy as np
import cv2
from tools import PlotTools, harris, sobel
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
img = cv2.imread('inputs/check.bmp')[:, :, 2]
grad_x, grad_y = sobel.compute_gradients(img)
img_sobel = sobel.get_sobel(grad_x, grad_y)
PlotTools.display_and_save(sobel.normalize(img_sobel), "out/sobel_check.bmp")
PlotTools.display_and_save(sobel.normalize(grad_x), "out/gradx_check.bmp")
PlotTools.display_and_save(sobel.normalize(grad_y), "out/grady_check.bmp")
```

```python
R = harris.harris_cuda_caller(grad_x, grad_y, wd=5, alpha=0.05)
PlotTools.display_img(R)
```

```python
img = cv2.imread('inputs/simA.jpg')[:, :, 2]
grad_x, grad_y = sobel.compute_gradients(img)
harris_img = harris.harris_cuda_caller(grad_x, grad_y, wd=5, alpha=0.05)
PlotTools.display_img(harris_img)
```

```python
maxi = np.max(img)
harris_img_filtered = np.vectorize(lambda x: x if (x > 0.2*maxi) else 0 )(harris_img)
PlotTools.display_img(harris_img_filtered)
```
